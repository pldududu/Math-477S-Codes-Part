```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters from the paper (interpreted)
beta = 0.25  # infection rate
gamma = 0.083  # recovery rate
alpha = 0.02  # death rate
v = 0.01  # reintegration rate

# Initial population
N = 50
S0 = 36
V0 = 6  # will not appear in ODE version explicitly
I0 = 8
R0 = 0
D0 = 0

# ODE system based on equations (4)-(7)
def svird_model(y, t, beta, gamma, alpha, v):
    S, I, R, D = y
    dSdt = -beta * S * I / N + v * R
    dIdt = beta * S * I / N - gamma * I - alpha * I
    dRdt = gamma * I - v * R
    dDdt = alpha * I
    return [dSdt, dIdt, dRdt, dDdt]

# Time grid from t=0 to t=50 (same as diagram)
t = np.linspace(0, 50, 501)

# Initial conditions vector
y0 = [S0, I0, R0, D0]

# Integrate the SVIRD equations over the time grid
sol = odeint(svird_model, y0, t, args=(beta, gamma, alpha, v))
S, I, R, D = sol.T

# V is fixed at 6 in this simulation
V = np.ones_like(t) * V0

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, S, label='Susceptible', color='dodgerblue')
plt.plot(t, V, label='Vaccinated', color='red')
plt.plot(t, I, label='Infected', color='orange')
plt.plot(t, R, label='Recovered', color='green')
plt.plot(t, D, label='Dead', color='gray')
plt.axvline(x=5, color='black', linestyle='--')  # t=5 marker
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Diagram of 50 Individuals (ODE Simulation, t=5 Marked)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#t=5 Low Vs High Vaccine Comparison
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters from the paper
beta = 0.25
gamma = 0.083
alpha = 0.02
v = 0.01
N = 50
t = np.linspace(0, 50, 501)

# SVIRD model definition
def svird_model(y, t, beta, gamma, alpha, v):
    S, I, R, D = y
    dSdt = -beta * S * I / N + v * R
    dIdt = beta * S * I / N - gamma * I - alpha * I
    dRdt = gamma * I - v * R
    dDdt = alpha * I
    return [dSdt, dIdt, dRdt, dDdt]

# Scenario 1 initial conditions
S0_1, V0_1, I0_1, R0_1, D0_1 = 36, 6, 8, 0, 0
y0_1 = [S0_1, I0_1, R0_1, D0_1]
sol1 = odeint(svird_model, y0_1, t, args=(beta, gamma, alpha, v))
S1, I1, R1, D1 = sol1.T
V1 = np.ones_like(t) * V0_1

# Scenario 2 initial conditions
S0_2, V0_2, I0_2, R0_2, D0_2 = 16, 30, 4, 0, 0
y0_2 = [S0_2, I0_2, R0_2, D0_2]
sol2 = odeint(svird_model, y0_2, t, args=(beta, gamma, alpha, v))
S2, I2, R2, D2 = sol2.T
V2 = np.ones_like(t) * V0_2

# Plotting side by side
fig, axs = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Scenario 1 plot
axs[0].plot(t, S1, label='Susceptible', color='dodgerblue')
axs[0].plot(t, V1, label='Vaccinated', color='red')
axs[0].plot(t, I1, label='Infected', color='orange')
axs[0].plot(t, R1, label='Recovered', color='green')
axs[0].plot(t, D1, label='Dead', color='gray')
axs[0].axvline(x=5, color='black', linestyle='--')
axs[0].set_title('Scenario 1: Low Vaccination (t=5 marked)')
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Count')
axs[0].legend()
axs[0].grid(True)

# Scenario 2 plot
axs[1].plot(t, S2, label='Susceptible', color='dodgerblue')
axs[1].plot(t, V2, label='Vaccinated', color='red')
axs[1].plot(t, I2, label='Infected', color='orange')
axs[1].plot(t, R2, label='Recovered', color='green')
axs[1].plot(t, D2, label='Dead', color='gray')
axs[1].axvline(x=5, color='black', linestyle='--')
axs[1].set_title('Scenario 2: High Vaccination (t=5 marked)')
axs[1].set_xlabel('Time')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
